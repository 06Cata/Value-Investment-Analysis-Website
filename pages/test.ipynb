```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 x 值範圍
x = np.linspace(-10, 10, 100)

# 計算對應的 y 值
y = x

# 繪製函數圖形
plt.plot(x, y, label='f(x) = x')

# 加上標籤
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = x')

# 加上圖例
plt.legend()

# 顯示圖形
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 x 值範圍
x = np.linspace(-10, 10, 100)

# 計算對應的 y 值
y = x**2

# 繪製函數圖形
plt.plot(x, y, label='f(x) = x^2')

# 加上標籤
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = x^2')

# 加上圖例
plt.legend()

# 顯示圖形
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 x 值範圍（不包括 x = 2）
x = np.linspace(-10, 1.9, 100)
x_right = np.linspace(2.1, 10, 100)

# 計算對應的 y 值
y = 1 / (x - 2)
y_right = 1 / (x_right - 2)

# 繪製函數圖形
plt.plot(x, y, label='f(x) = 1 / (x-2)', color='blue')
plt.plot(x_right, y_right, color='blue')

# 加上標籤
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = 1 / (x-2)')

# 加上圖例
plt.legend()

# 顯示圖形
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
